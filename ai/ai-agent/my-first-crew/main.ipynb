```python
from crewai import Agent, Task, Crew
```

```python
import os

os.environ["OPENAI_API_KEY"] = "sk-yourkeyhere"  # Replace with your OpenAI API key

os.environ["OPENAI_MODEL_NAME"] = "gpt-3.5-turbo"  # or gpt-4
```

```python
planner = Agent(
    role="An expert task planner.", # What is Role? The role defines the agent's expertise and behavior.
    goal="Create a detailed plan to build a simple web application using Flask.", # What is Goal? The goal defines the main objective the agent should achieve.
    backstory="You are a skilled software architect with experience in web development.", # What is Backstory? The backstory provides context about the agent's background and experience.
    allow_delegation=False,
    verbose=True
)
```

```python
writer = Agent(
    role="A senior software developer proficient in Python and Flask.",
    goal="Write clean, efficient, and well-documented code for a Flask web application based on the provided plan.",
    backstory="You have 10 years of experience in software development and have built numerous web applications using Flask.",
    allow_delegation=False,
    verbose=True
) 
```

```python
editor = Agent(
    role="A meticulous code reviewer and editor.",
    goal="Review and edit the Flask web application code to ensure it adheres to best practices, is free of bugs, and is well-documented.",
    backstory="You have a keen eye for detail and have spent years reviewing and improving code quality in various software projects.",
    allow_delegation=False,
    verbose=True
)
```

```python
plan = Task(
    agent=planner,
    description="Create a detailed plan to build a simple web application using Flask.",
    expected_output="A detailed plan for building a Flask web application with clear steps and implementation details."
)
```

```python
write = Task(
    agent=writer,
    description="Write the code for the Flask web application based on the provided plan.",
    expected_output="Clean, efficient, and well-documented code for the Flask web application."
)
```

```python
edit = Task(
    agent=editor,
    description="Review and edit the Flask web application code to ensure it adheres to best practices, is free of bugs, and is well-documented.",
    expected_output="A reviewed and edited version of the Flask web application code with improvements and documentation."
)
```

```python
crew = Crew(
    agents=[planner, writer, editor],
    tasks=[plan, write, edit],
    verbose=2
)
```

```python
result = crew.kickoff(inputs:={})
```
