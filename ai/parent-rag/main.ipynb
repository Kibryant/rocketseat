```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_community.document_loaders import PyPDFLoader
from langchain.retrievers import ParentDocumentRetriever
from langchain.storage import InMemoryStore
```

```python
EMBEDDINGS_MODEL_NAME = "text-embedding-3-small"

EMBEDDINGS = OpenAIEmbeddings(model=EMBEDDINGS_MODEL_NAME)

LLM_MODEL_NAME = "gpt-3.5-turbo"
LLM = ChatOpenAI(model_name=LLM_MODEL_NAME, max_tokens=500, temperature=0)
```

```python
# Load the PDF document
PDF_NAME = "example.pdf"  # Replace with your PDF file path

loader = PyPDFLoader(PDF_NAME, extract_images=False)

pages = loader.load_and_split()
```
